## Current setup

The existing setup of the LegoPress, based on the slides by Paul Bureth

```python
import numpy as np
```

### Motor Characteristics

```python
omega_nominal_motor = 1600 # rpm
omega_max_motor = 5000 # rpm

T_nominal_motor = 1.210 # Nm
T_stall_motor = 131 # Nm
```

### Linear Movement
![](./figures/linear-movement.png)

```python
R = 42.5 # Diameter of z3 gear

n1 = 60/15 # z2 has 60 teeth, z1 has 15 teeth
n2 = 2*np.pi/(np.pi*R) # [1/mm]

n_linear = n1 * n2 # [1/mm]
```

```python
v_nominal = 2*np.pi*omega_nominal_motor/(60*n_linear) # [mm/s]
v_max = 2*np.pi*omega_max_motor/(60*n_linear) # [mm/s]

F_nominal = T_nominal_motor*n_linear*1000 # [N]
F_max = T_stall_motor*n_linear*1000 # [N]
```

### Rotation Movement

![](./figures/rotation-movement.png)

```python
R1 = 12.5
R2 = 180

n_r = R2/R1

omega_nominal = np.rad2deg(2*np.pi*omega_nominal_motor/(60*n_r)) # [deg/s]
omega_max = np.rad2deg(2*np.pi*omega_max_motor/(60*n_r)) # [deg/s]

T_nominal = T_nominal_motor*n_r # [Nm]
T_max = T_stall_motor*n_r # [Nm]
```
